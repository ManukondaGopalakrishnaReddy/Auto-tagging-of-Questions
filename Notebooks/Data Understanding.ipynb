```python
category_subcategory_mapping = df.groupby('Category')['Sub Category'].unique()

# Display the mapping of subcategories under each category
for category, subcategories in category_subcategory_mapping.items():
    print(f"{category}: {subcategories}")
```

```python
# Count the occurrences of each taxonomy
category_counts = df['Bloom\'s Taxonomy'].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Question Counts by Taxonomy')
plt.xlabel('Taxonomy')
plt.ylabel('Question Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the occurrences of each category
category_counts = df['Category'].value_counts()

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Question Counts by Category')
plt.xlabel('Category')
plt.ylabel('Question Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(30, 15))
sns.countplot(data=df, x='Sub Category')
plt.title('Question Counts by Sub Category')
plt.xlabel('Sub Category')
plt.ylabel('Question Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Difficulty Level')
plt.title('Question Counts by Difficulty Level')
plt.xlabel('Difficulty Level')
plt.ylabel('Question Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a scatter plot of Bloom's Taxonomy vs. Difficulty Level
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Bloom\'s Taxonomy', y='Difficulty Level', hue='Category')
plt.title('Bloom\'s Taxonomy vs. Difficulty Level')
plt.xlabel('Bloom\'s Taxonomy')
plt.ylabel('Difficulty Level')
plt.show()
```
